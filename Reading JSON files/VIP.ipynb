```python
# data_load = json.load(data)

total_len = len(data_load['facets'])

timeseries_len = len(data_load['facets'][0]['timeSeries'])

#print(data_load['metadata']['accounts'][0])

#print(total_len)

#print(data_load['metadata']['contents']['timeSeries']['contents'][0]['attribute'])

 

host_name = []

app_name = []

data_center = []

metric_value = []

perc = []

beginTime = []

timeseries_endTimeSeconds_result = []

timeseries_inspectedcount_result = []

timeseries_accounts = []

metric_name = []

min_of_hour = []

source = []

 

for i in range(0, total_len):

    for j in range(0,timeseries_len):

        app_name.append(data_load['facets'][i]['name'][0])

        data_center.append(data_load['facets'][i]['name'][1])
        
        host_name.append(data_load['facets'][i]['name'][2])

        metric_value.append(str(list(data_load['facets'][i]['timeSeries'][j]['results'][0].values())[0]))
        
#         perc.append(str((list(data_load['facets'][i]['timeSeries'][j]['results'][0].values())[0])['99']))

        start_time_series = datetime.utcfromtimestamp(data_load['facets'][i]['timeSeries'][j]['beginTimeSeconds']).strftime('%Y/%m/%d %H:%M:%S')

        beginTime.append(str(datetime.strptime(start_time_series, '%Y/%m/%d %H:%M:%S')))

        mins = start_time_series[14:16]

        min_of_hour.append(mins)

        end_time_series = datetime.utcfromtimestamp(data_load['facets'][i]['timeSeries'][j]['endTimeSeconds']).strftime('%Y/%m/%d %H:%M:%S')

        timeseries_endTimeSeconds_result.append(datetime.strptime(end_time_series, '%Y/%m/%d %H:%M:%S'))

        timeseries_inspectedcount_result.append(data_load['facets'][i]['timeSeries'][j]['inspectedCount'])

        timeseries_accounts.append(data_load['metadata']['accounts'][0])

        #datetime.utcnow()-----> need to give while istering.

        metric_name.append(data_load['metadata']['contents']['timeSeries']['contents'][0]['attribute'])

        source.append("NewRelic")

#print(app_name)

 

def convert_list_to_dataframe():

    NewRelic_list = pd.DataFrame(

        {'app_name': app_name,

        'data_center': data_center,
         
        'host_name' : host_name,
         
#         '99_perc': perc,

        'vip_max': metric_value,

        'beginTime': beginTime,
         
        'endTime': timeseries_endTimeSeconds_result, 

        'min_of_hour': min_of_hour,

        'metric_name' : metric_name,

        'source' : source

        })

    return NewRelic_list

 

NewRelic_data = convert_list_to_dataframe()

print(NewRelic_data.head())
```
